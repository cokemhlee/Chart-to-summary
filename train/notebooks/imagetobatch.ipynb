```python
import os
os.getcwd()
import numpy as np
from PIL import Image
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
base_dir = 'C:\\projects\\chart_scrape\\chart\\' 
print(os.listdir(base_dir))
```

```python
img=[]
itera = 0
for i in os.listdir(base_dir):
    itera = 0
#     if i == "areaChart":
#         continue
#     if i == "histograms":
#         continue
#     print(i)
    for j in os.listdir(base_dir + i):
        try:
            img_path = base_dir + i + "\\" + j
            lalu = cv2.imread(img_path)
            
            img.append(np.array(cv2.resize(cv2.imread(img_path),(128,128))))
            print(np.array(lalu).shape)
            plt.imshow(np.squeeze(lalu))
            plt.show()
            break
            itera = itera + 1
            if itera % 64 == 0 :
                img = np.array(img)
                try:
                    os.mkdir("C:\\projects\\chart_scrape\\pachhes\\" + str(i) + "\\")
                except:
                    pass
                np.save("C:\\projects\\chart_scrape\\pachhes\\" + str(i) + "\\" + str(itera), img, allow_pickle=True, fix_imports=True)
                img = []
                print(itera)
        except Exception as ex:
            print(str(ex))
            print(i)
            print(j)
            continue
```
